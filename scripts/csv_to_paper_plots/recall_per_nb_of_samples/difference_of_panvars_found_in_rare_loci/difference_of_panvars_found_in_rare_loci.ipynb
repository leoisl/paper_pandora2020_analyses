```python
import pandas as pd
import numpy as np
```

# Data reading

```python
df_illumina = pd.read_csv("../leandro/df_illumina.R_data.csv")
df_illumina
```

```python
df_nanopore = pd.read_csv("../leandro/df_nanopore.R_data.csv")
df_nanopore
```

# Restricting to rare genes

```python
df_illumina_rare_genes = df_illumina[df_illumina.NUMBER_OF_SAMPLES <= 5]
df_illumina_rare_genes
```

```python
df_nanopore_rare_genes = df_nanopore[df_nanopore.NUMBER_OF_SAMPLES <= 5]
df_nanopore_rare_genes
```

# Summing the number of found panvars

```python
df_illumina_rare_genes_panvars_summed = \
    df_illumina_rare_genes.groupby(by=["tool_long_description", "tool"])\
    .agg({"nb_of_found_PanVar": np.sum})\
    .reset_index()
df_illumina_rare_genes_panvars_summed
```

```python
df_nanopore_rare_genes_panvars_summed = \
    df_nanopore_rare_genes.groupby(by=["tool_long_description", "tool"])\
    .agg({"nb_of_found_PanVar": np.sum})\
    .reset_index()
df_nanopore_rare_genes_panvars_summed
```

# Select the best values for each tool

```python
df_illumina_best_values = \
    df_illumina_rare_genes_panvars_summed.sort_values(by=["nb_of_found_PanVar"], ascending=False).\
    groupby(by=["tool"]).first()\
    .reset_index()
df_illumina_best_values
```

```python
df_nanopore_best_values = \
    df_nanopore_rare_genes_panvars_summed.sort_values(by=["nb_of_found_PanVar"], ascending=False).\
    groupby(by=["tool"]).first()\
    .reset_index()
df_nanopore_best_values
```

# Get how many panvars we have in total in rare genes, to be able to compute the recall

```python
pangenome_variations_per_nb_of_samples = pd.read_csv("../../pandora1_paper_analysis_output_20_way/technology_independent_analysis/pangenome_variants_vs_samples/pangenome_variations_per_nb_of_samples.csv")
pangenome_variations_per_nb_of_samples
```

```python
pangenome_variations_per_nb_of_samples_in_rare_genes = pangenome_variations_per_nb_of_samples[pangenome_variations_per_nb_of_samples.NUMBER_OF_SAMPLES <= 5]
pangenome_variations_per_nb_of_samples_in_rare_genes
```

```python
number_of_panvars_in_rare_genes = len(pangenome_variations_per_nb_of_samples_in_rare_genes)
display(number_of_panvars_in_rare_genes)
```

# Compute recall

```python
df_illumina_best_values["recall"] = df_illumina_best_values["nb_of_found_PanVar"] / number_of_panvars_in_rare_genes
df_illumina_best_values
```

```python
df_nanopore_best_values["recall"] = df_nanopore_best_values["nb_of_found_PanVar"] / number_of_panvars_in_rare_genes
df_nanopore_best_values
```

# Create report

```python
def extract_value(df, queried_tool, column):
    return df[df.tool==queried_tool][column].to_list()[0]

from collections import defaultdict
tech_to_tool_to_measure = defaultdict(lambda: defaultdict(lambda: defaultdict(float)))

for measure in ["nb_of_found_PanVar", "recall"]:
    for tool in ["pandora with denovo", "samtools", "snippy"]:
        tech_to_tool_to_measure["illumina"][tool][measure] = extract_value(df_illumina_best_values, tool, measure)
    for tool in ["pandora with denovo", "medaka", "nanopolish"]:
        tech_to_tool_to_measure["nanopore"][tool][measure] = extract_value(df_nanopore_best_values, tool, measure)

tech_to_tool_to_measure
```

```python
print("For illumina:")
for tool in ["samtools", "snippy"]:
    print(f"Pandora identified {int(tech_to_tool_to_measure['illumina']['pandora with denovo']['nb_of_found_PanVar'] - tech_to_tool_to_measure['illumina'][tool]['nb_of_found_PanVar'])} more pan vars than {tool}")
    print(f"Pandora had a recall {round((tech_to_tool_to_measure['illumina']['pandora with denovo']['recall'] - tech_to_tool_to_measure['illumina'][tool]['recall'])*100, 1)}% higher than {tool}")

print("For nanopore:")
for tool in ["medaka", "nanopolish"]:
    print(f"Pandora identified {int(tech_to_tool_to_measure['nanopore']['pandora with denovo']['nb_of_found_PanVar'] - tech_to_tool_to_measure['nanopore'][tool]['nb_of_found_PanVar'])} more pan vars than {tool}")
    print(f"Pandora had a recall {round((tech_to_tool_to_measure['nanopore']['pandora with denovo']['recall'] - tech_to_tool_to_measure['nanopore'][tool]['recall'])*100, 1)}% higher than {tool}")
```
